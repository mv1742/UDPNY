# Modelling Binary Variables using Census and Twitter data

This notebook contains the processing and modelling steps taken to measure the performance of Census + Twitter data models.

It contains several sections: 
   - Data Processing: Importing, cleaning, and merging all census and typology data.
   - Modelling Baseline Census: Creating and testing an initial census baseline using DT, RF, and Logit models.
   - Modelling Twitter Data Alone: A simple exercise to see the performance of Twitter data only in a Decision Tree model.
   - Modelling Twitter + Census: Combining Twitter and Census data in a single featureset to evaluate performance. 
   - Incorporating Distance Features: Adding in features engineered to measure distance to identified hotspots.
   - Incorporating Visitor Demographic & Timing: Adding in and assessing the final full set of Twitter + census features.

```python
def featureImportancePlot(rf, labels):
    '''plots feature importance for random forest
    rf: the random forest model fit to the data
    labels: the names of the features
    '''
    
    importances = rf.feature_importances_[:]
    indices = np.argsort(importances)[::-1]
    #std = np.std([tree.feature_importances_ for tree in rf.estimators_],
    #         axis=0)

    pl.figure(figsize=(10,5))
    pl.title("Feature importances")
    pl.bar(range(indices.shape[0]), rf.feature_importances_[indices],
       color="SteelBlue", #yerr=std[indices]
           align="center")
    pl.xticks(range(indices.shape[0]), np.array(labels)[indices], rotation=90)
    pl.ylabel('Weights')
    pl.xlabel('Features')
    pl.xlim([-1, indices.shape[0]])
    pl.show()
#featureImportancePlot(rf, X_foursquare.columns)
```
